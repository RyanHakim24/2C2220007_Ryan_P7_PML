```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

data = {
    'x' : [8, 3, 5, 7, 1, 2, 6, 7, 4, 9],
    'y' : [2, 2, 1.5, 1, 2.5, 3, 1.5, 2, 2, 1.5]
}

sks = pd.DataFrame(data)

sks
```

```python
sks['xy'] = sks['x'] * sks['y']
sks['x**2'] = sks['x'] ** 2
sks['y**2'] = sks['y'] ** 2

sks
```

```python
n = len(sks)

total = {
    'sum' : ['sum(x)', 'sum(y)', 'sum(xy)', 'sum(x^2)', 'sum(y^2)']
}
total2 = pd.DataFrame(total)

total2['Total'] = [sks['x'].sum(), sks['y'].sum(), sks['xy'].sum(), sks['x**2'].sum(), sks['y**2'].sum()]

total2
```

```python
r = (n * total2['Total'].iloc[2] - total2['Total'].iloc[0] * total2['Total'].iloc[1]) / np.sqrt((n * total2['Total'].iloc[3] - total2['Total'].iloc[0]**2) * (n * total2['Total'].iloc[4] - total2['Total'].iloc[1]**2))

r.round(3)
```
